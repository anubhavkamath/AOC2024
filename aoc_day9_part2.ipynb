```python
print('2024, Day 9 Part 2')
print('------------------------------------------')
from collections import deque
fpath = 'C:/Users/akamath/Documents/AOC2024/input9.txt'
file = open(fpath)
line = file.readline().strip()
queue = deque()
cnt = 0
total_nums = 0
file_list = []
def print_list(l):
    s = ''
    sum = 0
    for i in range(len(l)):
        num = l[i]
        if num != '.':
            sum += i * int(num)
        s = s+ str(l[i])
    
    print(l)
    print('Sum=', sum)

def get_block_info(num1, num2, id):
    global file_list
    nums = 0
    for i in range(int(num1)):
        file_list.append(id)
        nums += 1
    for j in range(int(num2)):
        file_list.append('.')
    return nums

def get_last_num_index(starting_index):
    # Item to exclude
    global file_list
    exclude_item = '.'
    
    # Find the last index of an item that is not 'exclude_item'
    found_index = None
    i = starting_index
    while(True):
        if file_list[i] == '.':
            i -=1
        else:
            found_index = i
            break
    return found_index

def get_space_index(start_index, need, num):
    global file_list
    num_spaces = 0
    for i in range(start_index, len(file_list)):
        if (file_list[i] == num):
            break
        if (start_index > file_list.index(num)):
            return
        if file_list[i] == '.':
            num_spaces += 1
            if num_spaces >= need:
                break
        else:
            num_spaces = 0
    if num_spaces >= need:   
        return i - num_spaces + 1
    return -1

def get_num_count(num):
    global file_list
    num_count = 0
    for item in reversed(file_list):
        if item != '.' and item == num:
            num_count += 1
    return num_count
        
def make_file_list():
    id = 0
    i = 0
    for c in line:
        queue.append(c)
    id = 0
    nums = 0
    while(True):
        num1 = None
        num1 = None
        if queue:
            num1 = queue.popleft()
        if queue:        
            num2 = queue.popleft()
        if num1:
            if not num2:
                num2 = 0
            nums += get_block_info(num1, num2, id)
        else:
            break
        id += 1
    return nums

def move_block_of_numbers(index, n, num):
    global file_list
    file_list = ['.' if item == num else item for item in file_list]
    file_list[index:index+n] = [num] * n
    return file_list
    
def rearrange_list(total_nums):
    global file_list
    last_num_index = get_last_num_index(len(file_list)-1)
    first_space_index = file_list.index('.')
    i = 0
    num_count = 0
    while(True):
        num = file_list[last_num_index]

        
        # the count of numbers at the end of the list
        num_count = get_num_count(num)
        first_space_index = file_list.index('.')

        # The number of available spaces in the beginning of the list
        space_index = get_space_index(first_space_index, num_count, num)
        if space_index and space_index >= 0:
            # insert the block of numbers in the spaces
            move_block_of_numbers(space_index, num_count, num)
            print('MOVED block: space_index:', space_index, 'Count:', num_count, 'Num:', num)
        i += 1
        if (last_num_index - num_count <= 0):
            break
        if i % 100 == 0:
            print('I:',i)
        last_num_index = get_last_num_index(last_num_index - num_count)


total_nums = make_file_list()
print_list(file_list)
rearrange_list(total_nums)
print_list(file_list)
# Answer is 6363913128533
# This took about a day and a half to complete running!
```
