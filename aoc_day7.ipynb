```python
path = 'C:/Users/akamath/Documents/AOC2024/input7.txt'
print('------------------------------------------')
print('2024, Day 7 Part 1 and 2')
print('------------------------------------------')
input = []
import ast
import re
from itertools import permutations, product
# perms = [''.join(p) for p in permutations('stack')]
permutations = []
result = 0
total_sum = 0
OPERS = '+*|'

def split_it(sstr):
    num_strs = re.split(r'[*+|\s]', sstr)
    return num_strs

#Takes a string and evalutes the expression and returns a string
# This function used for part 2 only
def calculate(calcstr):
    opers = [char for char in calcstr if char in OPERS]
    nums = split_it(calcstr)
    # for i in range(len(opers)):
    result = int(nums[0])
    for i in range(len(nums)-1):
        num = nums[i+1]
        if opers[i] == '+':
            result = result + int(num)
        elif opers[i] == '*':
            result = result * int(num) 
        elif opers[i] == '|':
            res = str(result)
            res += num
            result = int(res)
    return str(result)
    
def read_file():
    file = open(path)
    total_sum = 0
    while True:
        permutations = []
        line = file.readline().strip()
        if not line:
            break;
        entry = { 'num': int(line[:line.index(':')]), 'vals': list(map(int, line[line.index(':')+2:].split())) }
        perms = product(['+', '*'], repeat=len(entry.get('vals'))-1)

        permutations = [p for p in perms] # this will be a list of tuples containing permutations of operations
        got_match = False
        for ops in permutations:
            result = entry.get('vals')[0]
            for i in range(len(entry['vals'])-1):
                val = entry.get('vals')[i+1]
                if ops[i] == '+':
                    result = result + val
                elif ops[i] == '*':
                    result = result * val
            #print('Result:', result, entry.get('num'))
            if result == entry.get('num'):
                print('Got match:', entry.get('num'), result)
                total_sum += result
                got_match = True
                break
        perms = product(['+', '*', '|'], repeat=len(entry.get('vals'))-1)
        permutations = [p for p in perms] # this will be a list of tuples containing permutations of operations
        
        # This if block is for part2
        if not got_match:
            # For part 2 check if concatenation at any point between the vals results in num
            vals = entry.get('vals')
            num = entry.get('num')
            for ops in permutations:
                i = 0
                op_string = ''
                if '|' in ops:
                    for op in ops:
                        op_string += str(vals[i]) + op
                        i += 1
                    op_string += str(vals[i])
                    cnt = op_string.count('|')
                    result = calculate(op_string)
                        
                    if (int(result) == num):
                        total_sum += int(result)
                        print('Got match in concatenation:', num, total_sum)
                        got_match = True
                    if got_match:
                        break   
                i += 1                              
    file.close()
    print('Total sum:', total_sum) # part 2 20928985450275


read_file()
```
