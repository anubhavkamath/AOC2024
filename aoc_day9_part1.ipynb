```python
print('2024, Day 9 Part 1')
print('------------------------------------------')
from collections import deque
fpath = 'C:/Users/akamath/Documents/AOC2024/input9.txt'
file = open(fpath)
line = file.readline().strip()
queue = deque()
cnt = 0
file_list = []
total_nums = 0
def print_list(l):
    s = ''
    sum = 0
    for i in range(len(l)):
        num = l[i]
        if num == '.':
            break
        sum += i * int(num)
        s = s+ str(l[i])
    
    # print(l)
    print('Sum=', sum)

def get_block_info(num1, num2, id):
    nums = 0
    for i in range(int(num1)):
        file_list.append(id)
        nums += 1
    for j in range(int(num2)):
        file_list.append('.')
    return nums

def get_last_num_index():
    # Item to exclude
    exclude_item = '.'
    
    # Find the last index of an item that is not 'exclude_item'
    found_index = None
    for index in range(len(file_list) - 1, -1, -1):  # Iterate backwards through the list
        if file_list[index] != exclude_item:
            found_index = index
            break
    return found_index
        
def make_file_list():
    id = 0
    i = 0
    for c in line:
        queue.append(c)
    id = 0
    nums = 0
    while(True):
        num1 = None
        num1 = None
        if queue:
            num1 = queue.popleft()
        if queue:        
            num2 = queue.popleft()
        if num1:
            if not num2:
                num2 = 0
            nums += get_block_info(num1, num2, id)
        else:
            break
        id += 1
    return nums
        
def rearrange_list(total_nums):
    last_num_index = get_last_num_index()
    first_space_index = file_list.index('.')
    i = 0
    while(last_num_index):
        num = file_list[last_num_index]
        file_list[last_num_index] = '.'
        file_list[first_space_index] = num
        last_num_index = get_last_num_index()
        first_space_index = file_list.index('.')
        if (first_space_index >= total_nums-1):
            break


total_nums = make_file_list()
rearrange_list(total_nums)
print_list(file_list)
#00...111...2...333.44.5555.6666.777.888899
# 0099811188827773336446555566..............
# 00998111888277733364465555666666777.......
# 6340197768906
    
```
