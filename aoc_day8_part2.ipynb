```python
path = 'C:/Users/akamath/Documents/AOC2024/input8.txt'
file = open(path)
print('------------------------------------------')
print('2024, Day 8')
print('------------------------------------------')
coords = []
lines = file.readlines()
ant_dict = { }
valid_antinodes = []
extra = 0

# Convert coords back to string and print on separate line for readability    
def print_it():
    l = []
    cnt = 0
    s = ''
    for items in coords:
        s = ''
        for item in items:
            s += item  
        s + '\n'
        l.append(s)
    for item in l:
        print(item)

# Reads in the file
def read_it():
    for line in lines:
        line = line.strip()
        coords.append(list(line))
        
# adds key and coordinate to dict (creates it if it doesn't exist)
def add_to_dict(ant_key, coord_vals):
    # Check if the key exists, and if not, create an empty list
    if ant_key not in ant_dict:
        ant_dict[ant_key] = []
    ant_dict[ant_key].append(coord_vals)
        
# Finds the coordinates for all matching characters
def find_coords():
    # y is vertical coordiate but reverse of actual coordinate system, top is 0
    for y in range(len(coords)):
        # x increases from left to right like normal
        for x in range(len(coords[0])): 
            if coords[y][x] != '.':
                ant = coords[y][x]
                add_to_dict(ant, (y,x))

def is_in_bounds(antinode):
    if antinode[0] >= 0 and antinode[0] < len(coords) and antinode[1] >= 0 and antinode[1] < len(coords[0]):
        return True
    return False

def mark_antinodes():
    for antinode in valid_antinodes:
        coords[antinode[0]][antinode[1]] = 'X'

    
# Given coordinates for each antenna create all possible lines and calculate the two antinode points
# THis is way too verbose. Could have split this up into functions or implemented it better but it works
def find_antinodes():
    for ant, coords in ant_dict.items():
        for i in range(len(coords)-1):
            for j in range(i+1, len(coords)):
                if coords[j][1] - coords[i][1] == 0:
                    slope = 1
                else:
                    slope = (coords[j][0] - coords[i][0]) / (coords[j][1] - coords[i][1])
                coord1 = coords[i]
                coord2 = coords[j]
                ydiff = abs(coord2[0] - coord1[0])
                xdiff = abs(coord2[1] - coord1[1])
        
                if slope < 0:
                    ii = 1
                    ib1 = True
                    ib2 = True
                    while (True):
                        if not ib1 and not ib2:
                            break
                        ant1_y  = min(coord1[0], coord2[0]) - ydiff * ii
                        ant1_x = max(coord1[1], coord2[1]) + xdiff * ii
                        # Checks to see if it is valid
                        if is_in_bounds((ant1_y, ant1_x)):
                            if (ant1_y, ant1_x) not in valid_antinodes:
                                valid_antinodes.append((ant1_y, ant1_x))
                        else:
                            ib1 = False
                        ant2_y = max(coord1[0], coord2[0]) + ydiff * ii
                        ant2_x = min(coord1[1], coord2[1]) - xdiff * ii
                        if is_in_bounds((ant2_y, ant2_x)):
                            if (ant2_y, ant2_x) not in valid_antinodes:
                                valid_antinodes.append((ant2_y, ant2_x))
                        else:
                            ib2 = False
                        ii += 1
                elif slope > 0:
                    ii = 1
                    ib1 = True
                    ib2 = True
                    while (True):
                        if not ib1 and not ib2:
                            break
                            
                        ant1_y  = min(coord1[0], coord2[0]) - ydiff * ii
                        ant1_x = min(coord1[1], coord2[1]) - xdiff * ii
                        if is_in_bounds((ant1_y, ant1_x)):
                            if (ant1_y, ant1_x) not in valid_antinodes:
                                valid_antinodes.append((ant1_y, ant1_x))
                        else:
                            ib1 = False
                        ant2_y = max(coord1[0], coord2[0]) + ydiff * ii
                        ant2_x = max(coord1[1], coord2[1]) + xdiff * ii
                        if is_in_bounds((ant2_y, ant2_x)):
                            if (ant2_y, ant2_x) not in valid_antinodes:
                                valid_antinodes.append((ant2_y, ant2_x))
                        else:
                            ib2 = False
                        ii += 1
                elif slope == 0:
                    ii = 1
                    ib1 = True
                    ib2 = True
                    while (True):
                        if not ib1 and not ib2:
                            break
      
                        ant1_y  = coord1[0]
                        ant1_x = min(coord1[1], coord2[1]) - xdiff * ii
                        if is_in_bounds((ant1_y, ant1_x)):
                            if (ant1_y, ant1_x) not in valid_antinodes:
                                valid_antinodes.append((ant1_y, ant1_x))
                        else:
                            ib1 = False  
                        ant2_y = coord1[0]
                        ant2_x = max(coord1[1], coord2[1]) + xdiff * ii
                        if is_in_bounds((ant2_y, ant2_x)):
                            if (ant2_y, ant2_x) not in valid_antinodes:
                                valid_antinodes.append((ant2_y, ant2_x))
                        else:
                            ib2 = False
                        ii += 1
                else:
                    ii = 1
                    ib1 = True
                    ib2 = True
                    while (True):
                        if not ib1 and not ib2:
                            break
                        ant1_x  = coord1[1]
                        ant1_y = min(coord1[0], coord2[0]) - ydiff * ii
                        if is_in_bounds((ant1_y, ant1_x)):
                            if (ant1_y, ant1_x) not in valid_antinodes:
                                valid_antinodes.append((ant1_y, ant1_x))
                        else:
                            ib1 = False  
                        ant2_x = coord2[1]
                        ant2_y = max(coord1[0], coord2[0]) + ydiff * ii
                        if is_in_bounds((ant2_y, ant2_x)):
                            if (ant2_y, ant2_x) not in valid_antinodes:
                                valid_antinodes.append((ant2_y, ant2_x))
                        else:
                            ib2 = False
                        ii += 1
            
    not_in_antinodes = []
    cnt = 0
    # This last part is to count the  antennas that are not in valid_antinodes
    # If there are at least 3 of a type of antenna count them because they will
    # form a line with the others (minimum is 2 lines)
    for ant, coords in ant_dict.items():
        result = [item for item in coords if item not in valid_antinodes]
        if len(coords) >= 3:
            cnt += len(result)
    print('Count:', len(valid_antinodes) + cnt) # 1280 for part 2

read_it()
find_coords()
print_it()
find_antinodes()
mark_antinodes()
print_it()





```
