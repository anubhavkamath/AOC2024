```python
path = 'C:/Users/akamath/Documents/AOC2024/input5.txt'
file = open(path)
print('------------------------------------------')
print('2024, Day 5 Part 1')
print('------------------------------------------')
    
rules= []
inputs = []
while True:
    line = file.readline()
    if not line:
        break;
    line = line.strip()
    if '|' in line:
        rules.append(line)
    elif len(line) > 0:
        inputs.append(line)
rule_list = []
for rule in rules:
    t_rule = tuple(map(int, rule.split('|')))
    rule_list.append(t_rule)

def check_rule(num0,num1):
    broke_rule = False
    for rule in rule_list:
        if rule[0] == num1 and rule[1] == num0:
            print(rule, num0, num1)
            broke_rule = True
            break
    return broke_rule
    
def rules_check(inp):
    nums = list(map(int, inp.split(',')))
    failed = False
    print('Checking:', nums)
    for i in range(len(nums)-1):
        if check_rule(nums[i], nums[i+1]):
            failed = True
            break
    if not failed:
        return nums[len(nums)//2]
    return 0

sum = 0
for input in inputs:
    sum += rules_check(input)

print(sum) #4185
```
