```python
import re
path = 'C:/Users/akamath/Documents/AOC_2024/input3.txt'
file = open(path)
print('------------------------------------------')
print('2024, Day 2, part 1')
print('------------------------------------------')
safe = 0
line = ''
while True:
    l = file.readline()
    if not l: break
    line += l
# print(line)
total = 0
# find all patterns that match mul(x,x) where x can be 1-3 digits
pattern = re.compile("mul\((\d{1,3}),(\d{1,3})\)")
#For part 2
do_pattern = re.compile("do\(\)")
dont_pattern = re.compile("don\'t\(\)")
def get_total(match, line):
    tot = 0
    if match:
        nums = re.findall(r'\d+', match.group())
        print('nUMS:', nums)
        tot = int(nums[0]) * int(nums[1])
        line = line[match.end():]
    return tot, line
             
do_it = True
# at least one valid mul instruction has to be found first
match = pattern.search(line)
t, line = get_total(match, line)
total += t
# if total is > 0 means we found one, now apply the conditions for part 2
if total > 0:
    while True:
        # If we are not already in a don't condition, look for don'ts
        if do_it:
            # Look for don't or muls and see which one is first
            dont_match = dont_pattern.search(line)
            match = pattern.search(line)
            #If neither was found we are done
            if not dont_match and not match:
                break
            #if a don't was found set the flag and continue remove the text until the end of dont
            if dont_match and dont_match.start() < match.start():
                # Igore muls until a do is found
                do_it = False
                line = line[dont_match.end():]
                continue
        else:
            # We are in a don't condition, the only thing to do here is look for a do
            do_match = do_pattern.search(line)
            if do_match:
                line = line[do_match.end():]
                do_it = True
                continue
            else:
                # if we don't find a do, we are done
                break
        # if we get here we are in a do and have found a mul instruction in the match var which is set in the do_it block
        t, line = get_total(match, line)
        total += t
    
print(total) #188116424 for part 1, 104245808 for part 2
```
